```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import ast

# Step 1: 读取事务数据
transactions = []
with open('../..//dataset/frequent_mining_edges/frequent_mining_edges.txt', 'r') as file:
    for line in file:
        transaction = ast.literal_eval(line.strip())  # 安全地转换为列表
        transactions.append(transaction)
```

```python
import pandas as pd
# Step 2: 使用 TransactionEncoder 转换为布尔矩阵
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Step 3: 使用 apriori 算法找到频繁项集
min_support = 0.10  # 支持度阈值
min_confidence = 0.70  # 置信度阈值
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)  # 支持度阈值可调
# Step 4: 生成关联规则
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)  # 置信度阈值可调
```

```python
# Step 5: 格式化为指定格式
output = []
for _, row in rules.iterrows():
    lhs = frozenset(row['antecedents'])
    rhs = frozenset(row['consequents'])
    confidence = row['confidence']
    output.append((lhs, rhs, confidence))

# Step 6: 写入文件
with open('frequent_combinations_output.txt', 'w') as f:
    f.write(str(output))
```
