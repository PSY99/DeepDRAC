```python
import pandas as pd
train_pre_time = pd.read_csv('../..//src/runtime_analysis/train/train_pre_time.csv')
train_tune_time = pd.read_csv('../..//src/runtime_analysis/train/train_tune_time.csv')
# 按照alert_num合并两个df
merged_df = pd.merge(train_pre_time, train_tune_time, on='alert_num')
# 计算get_data_time+train_time+train_fine_tune_time
merged_df['total_time'] = merged_df['get_data_time'] + merged_df['train_time'] + merged_df['train_fine_tune_time']
# 重新命名get_data_time,train_time为pre_get_data_time,pre_train_time
merged_df.rename(columns={'get_data_time': 'pre_get_data_time', 'train_time': 'pre_train_time'}, inplace=True)
merged_df.to_csv('../..//src/runtime_analysis/train/train_time_summary.csv', index=False)
```
